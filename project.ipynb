```python
import requests
website_url = requests.get('https://github.com/BHARATW1993/mygit/blob/master/data%20capstone%20project1.csv').text
from bs4 import BeautifulSoup
soup = BeautifulSoup(website_url,'lxml')
print(soup.prettify())
```

```python
My_table = soup.find('table',{'class':'js-csv-data csv-data js-file-line-container'})
My_table
```

```python
print(My_table.tr.text)
```

```python
headers="State,Population2018"
```

```python
table1=""
for tr in My_table.find_all('tr'):
    row1=""
    for tds in tr.find_all('td'):
        row1=row1+","+tds.text
    table1=table1+row1[1:]
print(table1)
```
